## **Prediction using Supervised Machine learning**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1) Importing the Data

```python
data= "http://bit.ly/w-data"
im_data=pd.read_csv(data)
print ("Data has been imported")
im_data.head()
```

### 2) Plotting the data points of scores vs hours on a 2-D graph to find any relationship between the two variables

```python
im_data.plot(x="Hours",y="Scores",style="o")
plt.title("Hours vs Scores Graph")
plt.xlabel("Hours studied")
plt.ylabel("Percentage")
plt.show()
```

### Hence, we can observe that there is a positive relationship between the number of hours studied and scores obtained

### 3) We will divide our data into attributes and labels

```python
x=im_data.iloc[:,:1].values
```

### x represents the array of values in "Hours" column of the data 

```python
y=im_data.iloc[:,1].values
```

### y represents the array of values in "Scores" column of the data

### 4) Now, splitting the data into train and test sets using train_test_split() method of Scikit-Learn library

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### we have chosen 20% of the data for testing purpose and remaining 80% for training purpose

### 5) Training the model

```python
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
lm.fit(x_train,y_train)
print("Model has been trained")
```

### 6) Plotting the test data and Regression line

```python
#Regression line
line= lm.coef_*x + lm.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

### 7) Predicting through our trained algorithm

```python
print(x_test)
#testing data in hours
Y=lm.predict(x_test)
#Predicting scores of testing data of hours on the basis of our trained model
```

### 8) Comparing the actual values to our predicted values

```python
df=pd.DataFrame({"Actual":y_test,"Predicted":Y})
df
```

### Question : What will be the predicted score if a student studies for 9.25 hrs/ day?

```python
hours=9.25
pred = lm.predict([[hours]])
print ("No. of Hours:{}".format(hours))
print ("Predicted score:{}".format(pred[0]))
```

```python
from sklearn import metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test,Y))
```
